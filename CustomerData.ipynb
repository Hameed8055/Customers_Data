```python
#Visualization
```

```python
sns.histplot(df['Amount'], kde=True)  # `kde=True` adds a curve; optional
plt.title("Distribution of Amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Country', y='Amount', data=df)
plt.xticks(rotation=90)
```

```python
#Amout_VS_Company
```

```python
top_companies = df['Company'].value_counts().index[:10]
sns.boxplot(x='Company', y='Amount', data=df[df['Company'].isin(top_companies)])
plt.xticks(rotation=90)
```

```python
#Unsupervised_Learning

```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = df.select_dtypes(include=['float64', 'int64'])  # Or add engineered features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)
df['Segment'] = kmeans.fit_predict(X_scaled)

sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df['Segment'])
```
